# DATA PREPROCESSING

```python
df = df.drop(['CustomerId', 'Surname', 'RowNumber'], axis = 1)
print(df.columns)
```

```python
df.shape
```

```python
x = df.iloc[:,0:10]
y = df.iloc[:,10]
print(x.shape)
print(y.shape)
print(x.columns)
```

### ENCODING 

```python
x = pd.get_dummies(x)
x.head()
```

```python
x.shape
```

### Splitting the data to Train & Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```
